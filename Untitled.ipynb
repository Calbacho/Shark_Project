```python
def take_age(x):
    if '&' in x:
        lst = x.split('&')
    elif 'or' in x:
        lst = x.split('or')
    elif 'and' in x:
        lst = x.split('and')
    elif 'to' in x:
        lst = x.split('to')
    else:
        return x
    
    media = (int(lst[0])+int(lst[1]))//2
    return  str(media)

take_age('29&56')
```
